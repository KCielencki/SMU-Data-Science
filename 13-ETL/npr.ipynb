```python
url = 'https://www.npr.org/sections/news/'
```

```python
response = requests.get(url)
response.status_code
```

```python
html = response.text
html[0:50]
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
stories = []
headlines = soup.find_all('h2', class_='title')

for headline in headlines:
    stories.append(headline.text)
```

```python
lists = soup.find_all(class_="title")

href = []
urlMain = []

for headline in lists:
    href.append(headline.a.get("href"))
    urlMain.append(3)
```

```python
times = soup.find_all(class_="date")

date = []

for each in times:
    date.append(each.text.replace("",""))
```

```python
len(date)
```

```python
nprDf = pd.DataFrame()

nprDf["header"] = stories
nprDf["article_url"] = href
nprDf["site_id"] = urlMain
nprDf["date_published"] = date
```

```python
nprDf["tag_id"] = np.nan
nprDf
```

```python
filepath = r"C:\Users\kylec\Desktop\tags.csv"
tagsDf = pd.read_csv(filepath)
tagsDf = tagsDf.reset_index()
tagsDf.columns=["keyword_id", "keyword", "tag", "tag_id"]
tagsDf
```

```python
for i1, key in tagsDf.iterrows():
    for i2, headeline in nprDf.iterrows():
        if key.keyword in headeline.header.lower():
            nprDf.loc[i2, 'tag_id'] = key.tag_id
```

```python
nprDf = nprDf[["header", "article_url", "tag_id", "site_id", "date_published"]]
nprDf
```
